```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python

mnist=keras.datasets.fashion_mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data()
x_train,x_test= x_train/255.0,x_test/255.0
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[5])
```

```python
y_train[12]
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model=keras.Sequential([layers.Flatten(input_shape=(28,28)),layers.Dense(128,activation='relu'),layers.Dense(10,activation='softmax')])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=5)
```

```python
class_names=['T_shirt/top','Trowser','pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
num_classes=len(class_names)
print('number of classes:',num_classes)
```

```python
class_names={
    0:'T_shirt/top',
    1:'Trowser',
    2:'pullover',
    3:'Dress',
    4:'Coat',
    5:'Sandal',
    6:'Shirt',
    7:'Sneaker',
    8:'Bag',
    9:'Ankle boot'
}
```

```python
y_pred=model.predict(x_test)
```

```python
import numpy as np
y_pred1=np.argmax(y_pred,axis=1)
class_names[y_pred1[12]]
```

```python
class_names[y_test[12]]
```

```python
plt.imshow(x_test[5])
```

```python
plt.imshow(x_test[5],cmap='gray')
plt.title(f'y_test:{class_names[y_test[12]]},y_pred:{class_names[y_pred1[12]]}')
```

```python
import random
num_samples=10
selected_indices=random.sample(range(len(x_test)),num_samples)
for i in selected_indices:
  plt.figure()
  plt.imshow(x_test[i],cmap='gray')
  plt.title(f'True:{class_names[y_test[i]]},predicted:{class_names[y_pred1[i]]}')
  plt.show()
```
